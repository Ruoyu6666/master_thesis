# Eletricity

## Quantitles Regression

```python
# quantitles prediction
predictELprob = np.load("../predict/others/prob/predictELQ.npy")
# real value 
testELprob = np.load("../predict/others/prob/testELQ.npy")

print(predictELprob.shape)
print(testELprob.shape)
```

```python
l = 168

testELprob0 = np.zeros(l)       # real value
predictELprob0 = np.zeros(l)    # predict value
lowerELprob0 = np.zeros(l)      # 5
upperELprob0 = np.zeros(l)      # 95
lowerELprob02 = np.zeros(l)     # 25
upperELprob02 = np.zeros(l)     # 75

customerId = 65
for i in range(l):
    testELprob0[i]  = testELprob[i, customerId]             # real value
    predictELprob0[i]  = predictELprob[i,3, customerId,0]   # 50% quantitle -> mean  
    lowerELprob0[i] = predictELprob[i,0,customerId,0]       # 5%
    upperELprob0[i] = predictELprob[i,6,customerId,0]       # 95%

    lowerELprob02[i] = predictELprob[i,2,customerId,0]      # 25%
    upperELprob02[i] = predictELprob[i,4,customerId,0]      # 75%


x = list(range(0, l, 1))
```

```python
plt.figure(figsize=(9, 5))

plt.plot(x, predictELprob0, label='predict',color= "black")
plt.plot(x, testELprob0, label='real value',color= "crimson")
plt.plot(x, lowerELprob0, label='95% prediction interval', color= "royalblue")
plt.plot(x, upperELprob0, color= "royalblue")

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Eletricity')


plt.legend()
plt.fill_between(x, lowerELprob0, upperELprob0)
plt.grid(True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))


ax1.plot(x, predictELprob0, label='predict', color= "black")
ax1.plot(x, testELprob0, label='real value', color= "crimson")
ax1.plot(x, lowerELprob0, color= "royalblue")
ax1.plot(x, upperELprob0, label='95% prediction interval', color= "royalblue")
ax1.fill_between(x, lowerELprob0, upperELprob0)
ax1.set_ylabel('Value')
ax1.set_xlabel('Time')
ax1.set_title('fill between 5-')
ax1.legend()
ax1.grid(True)
#ax1.fill_between(x, lowerERprob02, upperERprob02, color= "C4")

# Start from 0
ax2.set_ylim(bottom=0)
ax2.plot(x, predictELprob0, label='predict', color= "black")
ax2.plot(x, testELprob0, label='real value',color= "crimson")
ax2.plot(x, lowerELprob0, color= "royalblue")
ax2.plot(x, upperELprob0, label='95% prediction interval',color= "royalblue")
ax2.fill_between(x, lowerELprob0, upperELprob0)
ax2.set_ylabel('Value')
ax2.set_xlabel('Time')
ax2.set_title('y axis starts from 0')
ax2.legend()
ax2.grid(True)
```
